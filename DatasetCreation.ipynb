```python
#Data set creation with scikit-lear
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=200, centers = 4, random_state = 42)

#Plot 
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap='viridis')
plt.show()
```

```python
#Another option
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
X_moons, y_moons = make_moons(n_samples=250, noise=0.1, random_state=42)

plt.scatter(X_moons[:,0], X_moons[:,1], c=y_moons, s=10, cmap='viridis')
plt.show()
```

```python
#Create a classification datasets in terms of decision boundaries
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X_circles, y_circles = make_circles(n_samples=250, factor=0.5, noise=0.1, random_state=42) 
plt.scatter(X_circles[:,0], X_circles[:,1], c=y_circles, s=60, cmap='viridis')
plt.show()
```

```python
#The number of decision boundaries can be increased, making the problem more complex
from sklearn.datasets import make_gaussian_quantiles
import matplotlib.pyplot as plt

X_gauss, y_gauss = make_gaussian_quantiles(n_samples=250, n_features=2, n_classes=4) 
plt.scatter(X_gauss[:,0], X_gauss[:,1], c=y_gauss, s=60, cmap='viridis')
plt.show()
```

```python
#A realistic case will be an imbalanced dataset
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
X_imb, y_imb= make_classification(n_samples=1500, n_features=20, n_informative=4, n_redundant=10, weights=[0.96, 0.04])

plt.scatter(X_imb[:,0], X_imb[:,1], c=y_imb, s=60, cmap='viridis')
plt.show()
```

```python
#Simulate a linear regression with noise
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np

X_reg, y_reg= make_regression(n_samples=400, n_features=1, noise=0.3)

dev = np.random.normal(loc=0, scale=20, size=y_reg.shape[0])
y_dev = y_reg + dev

plt.scatter(X_reg, y_dev, color = 'blue', s=30, marker='o', label='Regression data')
plt.show()
```
