```python
#Simple implementation of anomaly detection scenario
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create some sample data (1D array)
data = np.random.normal(scale=0.5, size=1000)
data=data.reshape(-1,1)
data
```

```python
plt.plot(range(len(data)), data)
plt.show()
```

```python
#Use scikit-learn to find anomalies
from sklearn.ensemble import IsolationForest 

#Create the model
clf =  IsolationForest(contamination=0.2) # Find 20% of anomalies of the given dataset
predictions = clf.fit_predict(data)
```

```python
#Anomalies are labeled ad -1 while normal values as 1
anomaly_indices = np.where(predictions==-1)[0]

plt.plot(range(len(data)), data)
plt.scatter(anomaly_indices, data[anomaly_indices], c='red')
plt.show()
```
